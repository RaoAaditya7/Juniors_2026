```python
#Q1.1
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
#q1.2
import math
a=int(input("Enter a number: "))
b=int(input("Enter another number: "))
primes=[]
for i in range(a,b+1):
    if i>1:
        for j in range(2,int(math.sqrt(i))+1):
            if i%j==0:
                break
        else:
            primes.append(i)
print("Prime numbers between",a,"and",b,":",primes)
```

```python
#q3.1 a)line plot
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = x**2

plt.figure()
plt.plot(x, y)
plt.title("Line Plot of f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
#q3.1 b)scatter plot
import numpy as np
import matplotlib.pyplot as plt
x = np.random.uniform(0, 1, 100)
y = np.random.normal(0.5, 0.1, 100)

plt.figure()
plt.scatter(x,y)
plt.title("Scatter Plot of Random Points")
plt.xlabel("x (Uniform Distribution)")
plt.ylabel("y (Normal Distribution)")
plt.show()
```

```python
#q3.1 c)histogram
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=1/1, size=500) 
plt.figure()
plt.hist(data, bins=30)
plt.title("Histogram of Exponential Distribution (λ = 1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [28, 34, 25, 42, 30],
    "Salary": [70000, 80000, 50000, 110000, 75000]
}
df = pd.DataFrame(data)
print(df)
df["Tax"] = df["Salary"]*0.2
print(df)
filtered_df = df[df["Age"] >= 30]
print(filtered_df)
avg_salary = df[df["Age"] < 30]["Salary"].mean()
print(avg_salary)
sorted_df = df.sort_values(by="Salary", ascending=False)
print(sorted_df)
```

```python
#q5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range(start="2025-01-01", end="2025-01-31")
t = np.arange(len(dates))
temperature = 15 + 10*np.sin(2*np.pi*t/31)


df = pd.DataFrame({"Date": dates,  "Temperature": temperature})

df.set_index("Date", inplace=True)

df["RollingAvg"] = df["Temperature"].rolling(window=7).mean()


plt.plot(df.index, df["Temperature"], label="Daily Temperature")
plt.plot(df.index, df["RollingAvg"], label="7-day Rolling Average")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("January 2025 Temperature Time Series")
plt.legend()
plt.show()

hot_days = df[df["Temperature"] > 20]
print("Days with temperature above 20°C:")
print(hot_days)
```
